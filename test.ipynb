```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
#from sklearn.metrics.pairwise import cosine_distances, euclidean_distances
from pylab import rcParams
from warping.dtw import *
```

```python
rcParams['figure.figsize'] = 15, 6
```

```python
n = 200
m = 10
x = np.sin(np.linspace(0,2*np.pi,n+m))*3 + np.random.rand(n+m)*0.5 + np.cos(np.linspace(0,5*np.pi,n+m))*2
y = np.sin(np.linspace(0,2*np.pi,n))*3 + np.random.rand(n)*0.5 + np.cos(np.linspace(0,5*np.pi,n))*1.5
print(x.shape)
print(y.shape)
```

```python
# Dynamic Time Warping
warp = dtw(x,y)
warp
```

```python
warp.plot()
```

```python
% timeit dtw(x,y)
```

```python
# Derivative DTW
warp = ddtw(x,y)
```

```python
warp
```

```python
warp.plot()
```

```python
% timeit ddtw(x,y)
```

```python
# For Multi Dimensional Arrays
n = 200
m = 10
c = 20
x = np.random.rand(n+m,c)
y = np.random.rand(n,c)
print(x.shape)
print(y.shape)
```

```python
warp = dtw(x,y,metric=euclidean_distances)
warp
```

```python
% timeit dtw(x,y,metric=euclidean_distances) # does not really slow down
```
