```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_distances, euclidean_distances
from pylab import rcParams
from pydtw.dtw import *
```

```python
rcParams['figure.figsize'] = 15, 6
```

```python
n = 200
m = 10
x = np.sin(np.linspace(0,2*np.pi,n+m))*3 + np.random.rand(n+m)*0.5 + np.cos(np.linspace(0,5*np.pi,n+m))*2
y = np.sin(np.linspace(0,2*np.pi,n))*3 + np.random.rand(n)*0.5 + np.cos(np.linspace(0,5*np.pi,n))*1.5
```

```python
print(x.shape)
print(y.shape)
```

```python
warp = dtw(x,y,metric=l1_norm)
```
